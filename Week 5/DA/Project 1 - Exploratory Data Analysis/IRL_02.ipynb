## Step-3:
Univariate analysis

```python
for i, column in enumerate(df.columns, 1):
    print(i, column)
```

```python
# plotting histograms with KDE
plt.figure(figsize = (25, 55))
plt.title("Histograms of numeric columns are shown")
for i, col, in enumerate(df.describe().columns, 1):
    plt.subplot(8, 3, i)
    sns.histplot(df[col], kde = True).set(title = col)
```

```python
# plotting PDF (as KDE)
plt.figure(figsize = (25, 55))
plt.title("PDF (as in KDE) of numeric columns are shown")
for i, col, in enumerate(df.describe().columns, 1):
    plt.subplot(8, 3, i)
    sns.kdeplot(df[col]).set(title = col)
```

```python
# plotting boxplots
plt.figure(figsize = (25, 55))
plt.title("Boxplots of numeric columns are shown")
for i, col, in enumerate(df.describe().columns, 1):
    plt.subplot(8, 3, i)
    sns.boxplot(y = df[col]).set(title = col)
```

```python
# plotting countplots for categorical data
plt.figure(figsize = (25, 55))
plt.title("Countplots categorical columns are shown")
for i, col, in enumerate(df.describe(include = "object").columns, 1):
    plt.subplot(3, 3, i)
    print(col)
    sns.countplot(df[col]).set(title = col)
```
