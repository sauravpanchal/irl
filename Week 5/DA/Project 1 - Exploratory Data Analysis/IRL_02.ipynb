```python
# detecting outliers with inter-quantile-range (IQR)
outliers = list()
def get_outlier_df(df, col):
    
    sorted(col) # sorting columns to get 25th and 75th percentile (like we do in median which is 50th percentile)
    Q1,Q3 = np.percentile(col , [25,75])
    IQR = Q3 - Q1
    lower = Q1 - (1.5 * IQR)
    upper = Q3 + (1.5 * IQR)
    outliers = df[(col < lower) | (col > upper)]

    return outliers

for numerical_column in df.describe().columns:
    outliers.append(len(get_outlier_df(df, df[numerical_column])))
# print(outliers)

outliers_stats_df = pd.DataFrame({
    "Column": df.describe().columns,
    "Number of outliers": outliers,
    "(%) forming of total data-points": list(map(lambda x: round((x / 3998) * 100, 2), outliers)),
})
outliers_stats_df
# l, u = get_outliers(df["Salary"])

# u = df.drop(t.index)
# sns.kdeplot(u["Salary"], color = "yellow", label = "No Outliers")
# sns.kdeplot(df["Salary"], color = "black", label = "With Outliers")
# plt.figure()
# sns.boxplot(y = u["Salary"], orient="v")
# # sns.boxplot(df["Salary"])
# plt.legend()
```

##### Q-2) Is there a relationship between gender and specialisation? (i.e. Does the preference of Specialisation depend on the Gender?)

```python
result = dict()
for val in df["Specialization"].values:
    male_count, female_count = 0, 0
    male_count = len(df[(df["Gender"] == "m") & (df["Specialization"] == val)])
    female_count = len(df[(df["Gender"] == "f") & (df["Specialization"] == val)])
    result[val] = {"m": male_count, "f": female_count}
# print(result)
# df["Specialization"].nunique()
```

```python
count = 0
# print("Specialization\t\t\t\t\tFemales\t\tMale")
for key, values in result.items():
    if result[key]["f"] >= result[key]["m"]:
        count += 1
        print(key)
print("\nAbove mentioned are the total {} Specializations female candidates are seems to be more indulged than male candidates.".format(count))
```

```python
gender_relation_df = pd.DataFrame(result)
gender_relation_df = gender_relation_df.transpose()
gender_relation_df.index.name = "Specialization"
gender_relation_df["(%) of female candidates"] = round((gender_relation_df["f"] / len(df["Gender"])) * 100, 2)
gender_relation_df
```
